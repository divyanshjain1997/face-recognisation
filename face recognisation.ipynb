# train a model to recognize

```python
import os
import numpy as np
import cv2
from PIL import Image

recog=cv2.face.LBPHFaceRecognizer_create()

face_classifier=cv2.CascadeClassifier(r"C:\Users\Dell\Anaconda3\pkgs\opencv-3.3.1-py36h20b85fd_1\Library\etc\haarcascades\haarcascade_frontalface_default.xml")

def getimawithid(path):
    imapath=[os.path.join(path,f) for f in os.listdir(path)]
    facesam=[]
    ids=[]
    for paths in imapath:
        pilimag=Image.open(paths).convert('L')
        npimag=np.array(pilimag,"uint8")
        #npimag=cv2.imread(paths,cv2.IMREAD_GRAYSCALE)
        faces=face_classifier.detectMultiScale(npimag)
        id=int(os.path.split(paths)[-1].split('.')[1])
        for (x,y,w,h) in faces:
            facesam.append(npimag[y:y+h,x:x+w])
            ids.append(id)
    return facesam,np.array(ids)

faces,ids=getimawithid(r"C:\Users\Dell\Desktop\dataset")
recog.train(faces,ids)
recog.write(r"C:\Users\Dell\Desktop\train\train.xml")
```
